```python
# 3. Online Shopping Cart System
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

class Cart:
    def __init__(self):
        self.items = {}

    def add_item(self, product):
        if product.stock <= 0:
            return f"{product.name} is out of stock."

        if product.name in self.items:
            if self.items[product.name]["qty"] >= product.stock:
                return "No more stock available."
            self.items[product.name]["qty"] += 1
        else:
            self.items[product.name] = {"price": product.price, "qty": 1}

        return f"{product.name} added to cart."

    def remove_item(self, product):
        if product.name not in self.items:
            return "Item not in cart."

        self.items[product.name]["qty"] -= 1

        if self.items[product.name]["qty"] <= 0:
            del self.items[product.name]

        return f"{product.name} removed from cart."

    def total_price(self):
        total = 0
        for item in self.items.values():
            total += item["price"] * item["qty"]
        return total

    def show_cart(self):
        if not self.items:
            return "Cart is empty."

        result = "Cart Items:\n"
        for name, data in self.items.items():
            result += f"{name} x{data['qty']} = {data['price'] * data['qty']}\n"

        result += f"Total: {self.total_price()}"
        return result


# Example usage
p1 = Product("Laptop", 1200, 5)
p2 = Product("Mouse", 50, 10)

cart = Cart()
print(cart.add_item(p1))
print(cart.add_item(p2))
print(cart.show_cart())
```

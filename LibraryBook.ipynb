```python
# 1. Library Book Management System
class Book:
    def __init__(self, title, author, status="available"):
        self.title = title
        self.author = author
        self.status = status

    def borrow(self):
        if self.status == "borrowed":
            return f"'{self.title}' is already borrowed."
        self.status = "borrowed"
        return f"You borrowed '{self.title}'."

    def return_book(self):
        if self.status == "available":
            return f"'{self.title}' is not borrowed."
        self.status = "available"
        return f"You returned '{self.title}'."

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, status="available"):
        self.books.append(Book(title, author, status))

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                return book.borrow()
        return "Book not found."

    def return_book(self, title):
        for book in self.books:
            if book.title == title:
                return book.return_book()
        return "Book not found."

    def show_books(self):
        result = "Library Books:\n"
        for book in self.books:
            result += f"{book.title} by {book.author} - {book.status}\n"
        return result


# Example usage
library = Library()
library.add_book("Python Basics", "Guido")
library.add_book("AI Guide", "John Doe")

print(library.borrow_book("Python Basics"))
print(library.return_book("Python Basics"))
print(library.show_books())
```

```python
# 2. Employee Payroll System
class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        if self.hours_worked <= 160:
            salary = self.hourly_rate * self.hours_worked
            overtime = 0
        else:
            overtime_hours = self.hours_worked - 160
            salary = self.hourly_rate * 160
            overtime = overtime_hours * (self.hourly_rate * 1.5)
            salary += overtime

        return salary, overtime

class Payroll:
    def __init__(self):
        self.employees = []

    def add_employee(self, name, rate, hours):
        if rate < 0 or hours < 0:
            return "Invalid input."
        self.employees.append(Employee(name, rate, hours))

    def show_payroll(self):
        if not self.employees:
            return "No employees found."

        report = ""
        for emp in self.employees:
            salary, overtime = emp.calculate_salary()
            report += f"\n{emp.name}: Hours={emp.hours_worked}, Salary={salary}, Overtime={overtime}"
        return report


# Example usage
payroll = Payroll()
payroll.add_employee("Justine", 100, 170)
payroll.add_employee("Kyle", 120, 150)

print(payroll.show_payroll())
```
